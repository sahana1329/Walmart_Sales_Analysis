```python
import pandas as pd
import sqlite3 # Or another driver like  pymysql

# Establish connection
conn = sqlite3.connect('walmart_data.db')

# Run SQL and load directly into a DataFrame
query = "SELECT * FROM walmart_clean_data WHERE rating > 9.0"
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
query = "SELECT count(*) FROM walmart_clean_data "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
query = "SELECT distinct payment_method, count(*) FROM walmart_clean_data group by payment_method "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
## Q1. find different payment method and number of transactions , number of qty sold 

query = "SELECT payment_method, count(*) as no_payments, sum(quantity) as no_qty_sold FROM walmart_clean_data group by payment_method "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
## Q2. identify the highest rated category in each branch, displaying the branch, category avg rating


query = "select * from (SELECT branch, category, avg(rating) as avg_rating, rank() over(partition by branch order by avg(rating) desc) as rank  FROM walmart_clean_data group by 1,2 ) where rank = 1 "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
# 3. Calculate the total quantity of items sold  per payment method. list payment_method and total_quantity




query = "SELECT payment_method, sum(quantity) as no_qty_sold FROM walmart_clean_data group by payment_method "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
 ## q4. Determine the average, minimum, and maximum rating of category for each city. list the city, average_rating, min_rating and max_rating.



query = "SELECT city, category, min(rating) as min_rating, max(rating) as max_rating, avg(rating) as avg_rating from walmart_clean_data group by 1, 2" 
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
## Q5. calculate the total profit for each category by considering total_profit as () unit_price * quantity * profit_margin). list category and total_profit, ordered from highest to lowest profit.




query = "select category, sum(total) as total_revenue, sum(total * profit_margin) as profit FROM walmart_clean_data group by 1  "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```

```python
## Q6. determine the most common payment method for each branch. display branch and the preferred_payment_method.




query = "WITH cte as (select branch, payment_method, count(*) as total_trans, rank() over(partition by branch order by count(*) desc) as rank FROM walmart_clean_data group by 1, 2) select * from cte where rank = 1  "
df = pd.read_sql_query(query, conn)
df_result = pd.read_sql_query(query, conn)

# Display the table
df_result
```
